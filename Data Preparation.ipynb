```python
import os
import os.path
import cv2
print("opencv version is:{}".format(cv2.__version__))
import pytesseract
from PIL import Image
import pandas as pd
from os import listdir
import seaborn as sns
import matplotlib as plt
from matplotlib import image
from matplotlib import pyplot
from numpy import asarray
import numpy as np
import tensorflow.keras as keras

from sklearn import preprocessing
```

```python
# Divide dataset into training (85%) and validation (15%)

df = pd.read_csv("/Users/kavya/PycharmProjects/Helicopter Airspeed Value Prediction/imagestimestamp.csv")
train_percent=.85
validate_percent=.15
seed = None
np.random.seed(seed)
perm = np.random.permutation(df.index)
m = len(df.index)
train_end = int(train_percent * m)
validate_end = int(validate_percent * m) + train_end
train = df.iloc[perm[:train_end]]
validate = df.iloc[perm[train_end:validate_end]]
test = df.iloc[perm[validate_end:]]

print(len(train))
print(len(validate))
```

```python
# Convert training data into csv 
train = train.filter(['Image Nmae', 'Airspeed Value'])
train= train.set_index('Image Nmae')
train.to_csv("/Users/kavya/PycharmProjects/Helicopter Airspeed Value Prediction/traindata.csv")

# Convert validation data into csv 
validate = validate.filter(['Image Nmae', 'Airspeed Value'])
validate= validate.set_index('Image Nmae')
validate.to_csv("/Users/kavya/PycharmProjects/Helicopter Airspeed Value Prediction/validatedata.csv")
```

```python
# Load data from csv
train_df = pd.read_csv("/Users/kavya/PycharmProjects/Helicopter Airspeed Value Prediction/traindata.csv")
valid_df = pd.read_csv("/Users/kavya/PycharmProjects/Helicopter Airspeed Value Prediction/validatedata.csv")

# Separate out our target values
y_train = train_df['Airspeed Value']
y_valid = valid_df['Airspeed Value']
del train_df['Airspeed Value']
del valid_df['Airspeed Value']


# Separate out our image vectors
x_train = train_df.values
x_valid = valid_df.values

end = preprocessing.LabelEncoder()
y_train = end.fit_transform(y_train)
y_valid = end.fit_transform(y_valid)


# Turn our scalar targets into binary categories
num_classes = 107 
y_train = keras.utils.to_categorical(y_train, num_classes)
y_valid = keras.utils.to_categorical(y_valid, num_classes)
```
